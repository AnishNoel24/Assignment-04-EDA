```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = pd.read_csv('/content/drive/MyDrive/DATASET/titanic.csv')
```

```python
# Display the first 5 rows of the dataset
print(df.head())
```

```python
# Get a summary of the dataset
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Fill missing values in 'Age' column with the median
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values in 'Embarked' column
df.dropna(subset=['Embarked'], inplace=True)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check for missing values in 'Cabin' column
print(df['Cabin'].isnull().sum())

# Replace missing values in 'Cabin' column with 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Plot a histogram of passenger ages
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the survival rate by passenger class
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Pclass', y='Survived', ci=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Plot a violin plot to show the relationship between passenger class, age, and survival
plt.figure(figsize=(10, 8))
sns.violinplot(data=df, x='Pclass', y='Age', hue='Survived', split=True)
plt.title('Passenger Class and Age vs. Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Plot the survival rate by sex
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Sex', y='Survived', ci=None)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Plot a count plot of passenger class
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Pclass')
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the survival rate by passenger class and sex
plt.figure(figsize=(10, 8))
sns.barplot(data=df, x='Pclass', y='Survived', hue='Sex', ci=None)
plt.title('Survival Rate by Passenger Class and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Plot a histogram of passenger fares
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Fare', bins=30, kde=True)
plt.title('Distribution of Passenger Fares')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
# Plot a box plot of age by passenger class
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Pclass', y='Age')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```
